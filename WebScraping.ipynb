## The implementation of (Book shop) scraping

```python
url='https://books.toscrape.com/catalogue/page-{}.html'
```

```python
page_number=1
while(page_number<=50):
  
    url = url.format(page_number)
    respose= rq.get(url)
    html_content=respose.content
    soup = BeautifulSoup(html_content, 'html.parser')
    books = soup.find_all('article',class_='product_pod')
    for i in range(len(books)):
        my_file=open('Book shop.txt','a')
        title=books[i].h3.a['title']
        price=books[i].find('p',class_='price_color').text
        rate=books[i].p['class'][1]
        print(f'Book title: {title}')
        print(f'Price: {price}')
        print(f'rating: {rate}')
        print("________________________________________")
        my_file.write(f'{title}\t{price}\t{rate}\n')
    page_number+=1
```

```python
import csv

filename = "Book shop.txt"

# Open the text file in read mode
with open(filename, 'r') as text_file:
    # Create a CSV writer object
    csv_writer = csv.writer(open('Book shop.csv', 'w', newline=''))

    # Read each line from the text file
    for line in text_file:
        # Split the line into fields if required
        fields = line.strip().split('\t')  # Assuming tab '\t' is the delimiter
        
        # Write the fields into the CSV file
        csv_writer.writerow(fields)

# Close the text file
text_file.close()

    
```
